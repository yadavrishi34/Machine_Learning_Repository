# Red Wine

The health benefits of red wine have been debated for some time.

Many believe that a glass each day is a valuable part of a healthy diet, while others think wine is somewhat overrated.

Studies have repeatedly shown that moderate red wine consumption seems to lower the risk of several diseases, including heart disease.

However, there is a fine line between moderate and excessive intake.


Red wine is made by crushing and fermenting dark-colored, whole grapes.

There are many types of red wine, which vary in taste and color. Common varieties include Shiraz, Merlot, Cabernet sauvignon, Pinot noir and Zinfandel.

The alcohol content usually ranges from 12–15%.

Consuming moderate amounts of red wine has been shown to have health benefits. This is mainly due to its high content of powerful antioxidants.

# Simple Linear Regression (SLR)

```python
# importing Liabraries

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from math import sqrt
```

```python
df.info()
```

```python
df
```

```python
# Selecting the input and output features for regression
# dividing array in x and y variable

# we can use directly by using arrary also, but i have taken here by using name.
# x=df.iloc[:,:-2].values
# y=df.iloc[:,-2].values)

features=['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides',
            'free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol']
target=['quality']

x=df[features]
y=df[target]
```

```python
# Dividing data in to train and test dataset in the ratio of 70:30 ratio using sklear train test split function

x_train,x_test,y_train,y_test=train_test_split(df,y,test_size=0.30,random_state=200)
```

```python
print(x_train.shape)
```

```python
print(x_test.shape)
```

```python
# fit on training set
Lin=LinearRegression()
Lin.fit(x_train,y_train)
```

```python
# Predict on test data

predict=Lin.predict(x_test)
```

```python
# cross check with predict data and test data
print(predict[:5])
print('-'*50)
print(y_test[:5])
```

```python
y_test.describe()
```

```python
Lin.score(x_train,y_train)
```

```python
Lin.coef_
```

```python
# printing result Error

print('error:')

# printing of MAE (Mean Absolute Error)

print(metrics.mean_absolute_error(y_test,predict))

# printing of MSE (Mean Square Error)

print(metrics.mean_squared_error(y_test,predict))

# printing RMSE (Root Mean Square Error)

print(np.sqrt(metrics.mean_squared_error(y_test,predict)))
```

# Decision Tree 

```python
# import liabraries

from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
dtr=DecisionTreeRegressor(random_state=0)
dtr.fit(x_train,y_train)
```

```python
predict=dtr.predict(x_test)
predict[:5]
```

```python
y_test[:5]
```

```python
# Evaluating Decision Tree Regression accuracy

print(np.sqrt(metrics.mean_squared_error(y_test,predict)))
```

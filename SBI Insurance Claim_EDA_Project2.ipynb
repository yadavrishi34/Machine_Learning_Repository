Data Description to understand

age : age of policyholder 

sex: gender of policy holder (female=0, male=1)

bmi: Body mass index, providing an understanding of body, weights that are relatively high or low relative to height, objective index of body weight (kg / m ^ 2) using the ratio of height to weight, ideally 18.5 to 25 steps: average walking steps per day of policyholder 

children: number of children / dependents of policyholder 

smoker: smoking state of policyholder (non-smoke=0;smoker=1) 

region: the residential area of policyholder in the US (northeast=0, northwest=1, southeast=2, southwest=3) 

charges: individual medical costs billed by health insurance insuranceclaim: yes=1, no=0

```python
# import alogrithm from sk learn

import sklearn

from sklearn.linear_model import LinearRegression

from sklearn import metrics

from sklearn.model_selection import train_test_split
```

```python
x=df.iloc[:,-2]
```

```python
x
```

```python
y=df.iloc[:,0]
```

```python
y
```

```python
plt.xlabel('charges')
plt.ylabel('age')
plt.scatter(x,y,color='blue')
plt.show()
```

```python
x_trains,x_test,y_trains,y_test=train_test_split(x,y,test_size=.30,random_state=45)
```

```python
x_trains.shape
```

```python
y_trains.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
# Reshaping Train data set in one column

x_trains=x_trains.values
x_trains=x_trains.reshape(-1,1)
x_trains.shape
```

```python
y_trains=y_trains.values
y_trains=y_trains.reshape(-1,1)
y_trains.shape
```

```python
x_test=x_test.values
x_test=x_test.reshape(-1,1)
x_test.shape
```

```python
y_test=y_test.values
y_test=y_test.reshape(-1,1)
y_test.shape
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_trains,y_trains)
```

```python
# predict the value

pred=lr.predict(x_test)
print("prediction:",pred)
```

```python
lr.score(x_trains,y_trains)
```

```python
plt.scatter(x_test,pred)
plt.show()
```

```python
lr.coef_
```

```python
# printing result Error

print('error:')

# printing of MAE (Mean Absolute Error)

print(metrics.mean_absolute_error(y_test,pred))

# printing of MSE (Mean Square Error)

print(metrics.mean_squared_error(y_test,pred))

# printing RMSE (Root Mean Square Error)

print(np.sqrt(metrics.mean_squared_error(y_test,pred)))
```
